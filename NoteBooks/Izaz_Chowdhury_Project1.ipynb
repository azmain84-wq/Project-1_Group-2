**Case 1 —** <span style="color: rgb(0, 0, 0); font-family: -webkit-standard; font-size: medium; caret-color: rgb(0, 0, 0);">The Late Shipment<br></span><span style="caret-color: rgb(0, 0, 0); color: rgb(0, 0, 0); font-family: -webkit-standard; font-size: medium;">A long-time customer claims their order arrived days after the promised delivery date. The warehouse insists everything went out on time, but the truth lies in the data.</span>  
<span style="caret-color: rgb(0, 0, 0); color: rgb(0, 0, 0); font-family: -webkit-standard; font-size: medium;">Can you find which orders were<span class="Apple-converted-space">&nbsp;</span></span> <span data-start="524" data-end="572" style="caret-color: rgb(0, 0, 0); color: rgb(0, 0, 0);">completed after their expected delivery date</span><span style="caret-color: rgb(0, 0, 0); color: rgb(0, 0, 0); font-family: -webkit-standard; font-size: medium;">?</span><span style="color: rgb(0, 0, 0); font-family: -webkit-standard; font-size: medium; caret-color: rgb(0, 0, 0);"><br></span>

```sql


SELECT TOP (20)
    o.OrderID,
    c.CustomerName,
    o.OrderDate,
    o.ExpectedDeliveryDate,
    o.PickingCompletedWhen
FROM Sales.Orders AS o
JOIN Sales.Customers AS c
    ON o.CustomerID = c.CustomerID
WHERE o.PickingCompletedWhen > o.ExpectedDeliveryDate
ORDER BY o.PickingCompletedWhen DESC;

```

\*\*Case 2 — The Silent Salesperson  
<span style="caret-color: rgb(0, 0, 0); color: rgb(0, 0, 0); font-family: -webkit-standard; font-size: medium;">Rumors are spreading that one salesperson has been quietly coasting through the quarter without pulling their weight. Management wants proof before the next review.</span>  
<span style="caret-color: rgb(0, 0, 0); color: rgb(0, 0, 0); font-family: -webkit-standard; font-size: medium;">Can you find which<span class="Apple-converted-space" style="">&nbsp;</span></span> <span data-start="373" data-end="419" style="caret-color: rgb(0, 0, 0); color: rgb(0, 0, 0);">salespeople have handled the fewest orders</span> <span style="caret-color: rgb(0, 0, 0); color: rgb(0, 0, 0); font-family: -webkit-standard; font-size: medium;"><span class="Apple-converted-space" style="">&nbsp;</span>in the system?</span>

```sql
SELECT TOP (20)
    e.PersonID,
    e.FullName,
    COUNT(o.OrderID) AS OrdersHandled
FROM Application.People AS e
LEFT JOIN Sales.Orders AS o
    ON e.PersonID = o.SalespersonPersonID
WHERE e.IsSalesperson = 1
GROUP BY e.PersonID, e.FullName
ORDER BY OrdersHandled ASC;

```

<span style="caret-color: rgb(0, 0, 0); color: rgb(0, 0, 0); font-family: -webkit-standard; font-size: medium;">Case 3 - The Vanishing Supplier<br></span><span style="caret-color: rgb(0, 0, 0); color: rgb(0, 0, 0); font-family: -webkit-standard; font-size: medium;">Several suppliers claim they’ve been sending shipments regularly, yet the warehouse hasn’t logged any recent orders from them. Something doesn’t add up.</span>  
<span style="caret-color: rgb(0, 0, 0); color: rgb(0, 0, 0); font-family: -webkit-standard; font-size: medium;">Can you find which<span class="Apple-converted-space" style="">&nbsp;</span></span> <span data-start="323" data-end="367" style="caret-color: rgb(0, 0, 0); color: rgb(0, 0, 0);">suppliers have no recent purchase orders</span> <span style="caret-color: rgb(0, 0, 0); color: rgb(0, 0, 0); font-family: -webkit-standard; font-size: medium;"><span class="Apple-converted-space">&nbsp;</span>or haven’t supplied anything at all?</span><span style="caret-color: rgb(0, 0, 0); color: rgb(0, 0, 0); font-family: -webkit-standard; font-size: medium;">.</span>

```sql
SELECT TOP (30)
    s.SupplierID,
    s.SupplierName,
    MAX(poh.OrderDate) AS LastOrderDate
FROM Purchasing.Suppliers AS s
LEFT JOIN Purchasing.PurchaseOrders AS poh
    ON s.SupplierID = poh.SupplierID
GROUP BY s.SupplierID, s.SupplierName
HAVING MAX(poh.OrderDate) < '2015-01-01'
    OR MAX(poh.OrderDate) IS NULL
ORDER BY LastOrderDate;
```

**Case 4 — The Duplicate Customer**

<span style="caret-color: rgb(0, 0, 0); color: rgb(0, 0, 0); font-family: -webkit-standard; font-size: medium;">Customer service keeps getting calls from what seem to be the same people under slightly different names. It looks like the system may contain duplicates.</span>  
<span style="caret-color: rgb(0, 0, 0); color: rgb(0, 0, 0); font-family: -webkit-standard; font-size: medium;">Can you find<span class="Apple-converted-space">&nbsp;</span></span> <span data-start="322" data-end="367" style="caret-color: rgb(0, 0, 0); color: rgb(0, 0, 0);">customers who share the same phone number</span><span style="caret-color: rgb(0, 0, 0); color: rgb(0, 0, 0); font-family: -webkit-standard; font-size: medium;">, revealing possible duplicate accounts?</span>

```sql
SELECT TOP (20)
    c.PhoneNumber,
    COUNT(c.CustomerID) AS DuplicateCount,
    STRING_AGG(c.CustomerName, ', ') AS Customers
FROM Sales.Customers AS c
WHERE c.PhoneNumber IS NOT NULL
GROUP BY c.PhoneNumber
HAVING COUNT(c.CustomerID) > 1
ORDER BY DuplicateCount DESC;
```

**Case 5 — The Overstocked Warehouse  
**<span style="caret-color: rgb(0, 0, 0); color: rgb(0, 0, 0); font-family: -webkit-standard; font-size: medium;">The warehouse manager complains that storage is full, yet sales haven’t increased. Could some products be sitting unsold for months?</span>  
<span style="caret-color: rgb(0, 0, 0); color: rgb(0, 0, 0); font-family: -webkit-standard; font-size: medium;">Find which products have<span class="Apple-converted-space">&nbsp;</span></span> <span data-start="447" data-end="511" style="caret-color: rgb(0, 0, 0); color: rgb(0, 0, 0);">high stock levels but haven’t appeared in any recent orders.</span>

```sql
SELECT TOP (20)
    p.StockItemID,
    p.StockItemName,
    h.QuantityOnHand,
    MAX(ol.LastEditedWhen) AS LastSoldDate
FROM Warehouse.StockItems AS p
JOIN Warehouse.StockItemHoldings AS h
    ON p.StockItemID = h.StockItemID
LEFT JOIN Sales.OrderLines AS ol
    ON p.StockItemID = ol.StockItemID
GROUP BY p.StockItemID, p.StockItemName, h.QuantityOnHand
ORDER BY h.QuantityOnHand DESC;


```

## **Case 6 — The Late Invoice (Final Version)**

Some invoices seem to take longer to confirm than others. Could delays in processing be hiding a bigger problem?  
Find how many days it took to confirm each invoice, starting with the slowest ones.

```sql
SELECT TOP (20)
    i.InvoiceID,
    c.CustomerName,
    i.InvoiceDate,
    i.ConfirmedDeliveryTime,
    DATEDIFF(DAY, i.InvoiceDate, i.ConfirmedDeliveryTime) AS DelayDays
FROM Sales.Invoices AS i
JOIN Sales.Customers AS c
    ON i.CustomerID = c.CustomerID
ORDER BY DelayDays DESC;




```

## **Case 7 — The Top Earning Country**

The CEO wants to know which countries bring in the most money. Are sales dominated by one region, or are profits spread evenly?  
Find the total revenue generated by each country, and show the top performers.

```sql
SELECT TOP (20)
    co.CountryName,
    SUM(il.ExtendedPrice) AS TotalRevenue
FROM Sales.InvoiceLines AS il
JOIN Sales.Invoices AS i
    ON il.InvoiceID = i.InvoiceID
JOIN Sales.Customers AS c
    ON i.CustomerID = c.CustomerID
JOIN Application.Cities AS city
    ON c.DeliveryCityID = city.CityID
JOIN Application.StateProvinces AS sp
    ON city.StateProvinceID = sp.StateProvinceID
JOIN Application.Countries AS co
    ON sp.CountryID = co.CountryID
GROUP BY co.CountryName
ORDER BY TotalRevenue DESC;

```
